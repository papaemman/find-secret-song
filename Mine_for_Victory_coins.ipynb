# Collect Metal song titles

## Setup

```python
# Load libraries
import pandas as pd
import re
import requests
```

## Website Scrapping

### Site 1

* https://www.sporcle.com/games/metalwolf123/top-1000-heavy-metal-songs-200-101/results

```python
# import requests
songs = []

for page_id in range(24):
    URL = f"https://forum.rpg.net/index.php?threads/what-are-the-1000-greatest-metal-songs-these-are.678171/page-{page_id}"
    page = requests.get(URL)

    # print(page.text)
    # type(page.text)
    songs.extend(re.findall( r'\d+\. (.*?)<', page.text))
    
```

```python
songs
```

```python
print("Total songs:", len(songs))
```

```python
# Clean up
songs = filter(lambda x: "-" in x, songs)
songs_clean = list(map(lambda x: x.split("- ")[1].lower().replace("'", "").replace(" ", "_"), songs))
songs_clean
```

```python
# Store songs
with open('data/songs.txt', 'a') as fp:
    fp.write('\n'.join(songs_clean))
```

# Try out all metal songs

https://victory.jon.io/claim/name_of_the_song

**Responses**

* 1xx - informational
* 2xx - success
* 3xx - redirection
* 4xx - client error
* 5xx - server error

```python
# Store songs
with open('data/songs.txt', 'r') as fp:
    songs_clean = fp.readlines()
```

```python
# More cleanup
songs_clean = list(map(lambda x: x.replace("\n", "").replace("&quot;",""), songs_clean))
```

```python
print("Total songs:", len(songs_clean))
```

```python
songs_clean
```

```python
for song_name in songs_clean:
    
    # song_name = "theroom"
    
    # Construct the url
    url_to_try = f"https://victory.jon.io/claim/{song_name}"
    
    # Print url
    print("URL:", end=" ")
    print(url_to_try)

    # Send a Get request
    r = requests.get(url_to_try)

    # Transform respond to string
    string_var = str(r.content, 'utf-8')


    # When url doesn't redirects to https://victory.jon.io/, this means that I have found the correct song.
    # How to find the differenciating part between responses? Compare html elements

    correct_song = 'content="https://victory.jon.io/"' not in string_var
    if correct_song:
        print("You find the correct song: ", song_name)
        break 
```

```python
##### Sanity check (try an existing link)
song_name = "theroom"
url_to_try = f"https://victory.jon.io/claim/{song_name}"

# Print url
print("URL:", end=" ")
print(url_to_try)

# Send a Get request
r = requests.get(url_to_try)

# Transform respond to string
string_var = str(r.content, 'utf-8')


# When url doesn't redirects to https://victory.jon.io/, this means that I have found the correct song.
# How to find the differenciating part between responses? Compare html elements

correct_song = 'content="https://victory.jon.io/"' not in string_var
if correct_song:
    print("You find the correct song: ", song_name)
```
